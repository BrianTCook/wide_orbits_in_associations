```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

cases = [ 3., 5., 2., 3., 3., 5., 5., 0.,
          8., 13., 33., 51., 48., 24., 58. ]

days = np.arange(0, len(cases))

ahh, bhh = (curve_fit(lambda t,a,b: a*np.exp(b*t),
                               days,  cases,  p0=(4, 0.1)))

print(ahh)

fit_x = np.linspace(0., 14., 100)
fit_y = [ ahh[0] * np.exp(ahh[1] * x) for x in fit_x ]

plt.scatter(days, cases, s=8, marker='x', c='r', label='reported cases')
plt.plot(fit_x, fit_y, c='k', label=r'$y = %.02f * \exp(%.02f * x)$'%(ahh[0], ahh[1]))
plt.legend(loc='best', fontsize=12)
plt.xlabel('Days since August 12', fontsize=14)
plt.ylabel('Positive COVID-19 Tests', fontsize=14)
plt.title('Exponential Growth, Ge')
```
